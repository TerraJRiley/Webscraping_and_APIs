```python
spell_window = open_browser()
```

```python
spell_window.get('https://dnd5e.fandom.com/wiki/List_of_Spells')
```

```python
spell_soup = BeautifulSoup(spell_window.page_source)
```

```python
list_of_dicts = []
for table in spell_soup.find_all('table'):#[0:1]:
    columns = [word.text.strip('\n').strip(' ') for word in table.find('thead').find_all('th')]
    print(columns)
    for row in table.find('tbody').find_all('tr'):#[1:3]:
        row_dict = {}
        cells_in_row = [cell.contents for cell in row.find_all('td')]
        #print(cells_in_row)
        for item_num in range(len(columns)):
            row_dict[columns[item_num]] = cells_in_row[item_num][0]
        list_of_dicts.append(row_dict)
            
        #for cell in cells_in_row:
        #    print(cell.contents)
        #print('------')
```

```python
spells_df = pd.DataFrame(list_of_dicts)
```

```python
def map_function(x):
    try:
        return x.attrs['href']
    except:
        return 'No link'
spells_df['href_link'] = spells_df['Name'].map(lambda x: map_function(x))
```

```python
spells_df['Name_html'] = spells_df['Name']
spells_df['Name'] = spells_df['Name'].map(lambda x: x.text)
```
